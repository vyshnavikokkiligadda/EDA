**Data Transformations**

- Generally data has  types of distributions

    - Normal distribution
    
    - Positive Skewed
    
    - Negative Skewed
    
- All the maths concepts developed by assumption as Data follows Normal distribution

- Data skew happens because of many reasons, in that one reason is outliers

- Oncee we perform the outlier analysis then we apply transformation methods, if the data does not follows normality

- We have some transformation methods mentioned below

    - Log Transformation
    
    - Exponential Transformation
    
    - Reciprocal Transformation
    
    - Square Transformation
    
    - Power Transormation

$Step_4$

**Log transformation**

- Log transformation means performing logaritham operations on original data

- It is an approach to convert into Normal distribution

- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logaritham')
plt.legend()
plt.show()
```

**Reciprocal Transformations**

- Reciprocal will fail when values are low

- For examples x=0 then 1/0 fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal')
plt.show()
```

**Square root transformation**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data, bins=20, label='Square_Root_Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
log_data=np.log(exp_data)
plt.subplot(1,3,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
log_data=np.log(exp_data)
plt.subplot(1,3,2).hist(log_data,bins=30,label='Logaritham')
plt.legend()
square_root_data=np.sqrt(exp_data)
plt.subplot(1,3,3).hist(square_root_data,bins=30,label="Square")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='box-cox',standardize=False)

X_Transformed = pt.fit_transform(exp_data.reshape(-1, 1))
print(X_Transformed)
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans = power.fit_transform(exp_data1)
plt.hist(data_trans)
```
